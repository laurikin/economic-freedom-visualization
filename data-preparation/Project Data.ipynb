```python
import pandas as pd
import numpy as np
```

```python
efi_raw = pd.read_csv('EconomicFreedomIndex.csv', delimiter=";")
gdp_raw = pd.read_csv('GDP_per_capita_2010$.csv', delimiter=";")
```

```python
gdp = gdp_raw.drop(columns=['Indicator Name', 'Indicator Code', 'Country Code', 'Unnamed: 64', '2019', '2018', '1960', '1961', '1962', '1963', '1964', '1965', '1966', '1967', '1968', '1969', '1971', '1972', '1973', '1974', '1976', '1977', '1978', '1979', '1981', '1982', '1983', '1984', '1986', '1987', '1988', '1989', '1991', '1992', '1993', '1994', '1996', '1997', '1998', '1999'])
gdp.head()
```

```python
gdp['Country Name'] = gdp['Country Name'].replace('Bahamas, The', 'Bahamas')
gdp['Country Name'] = gdp['Country Name'].replace('Congo, Rep.', 'Congo, Rep. of')
gdp['Country Name'] = gdp['Country Name'].replace('Czech Republic', 'Czech Rep.')
gdp['Country Name'] = gdp['Country Name'].replace('Slovak Republic', 'Slovak Rep')
gdp['Country Name'] = gdp['Country Name'].replace('Venezuela, RB', 'Venezuela')
gdp['Country Name'] = gdp['Country Name'].replace('Hong Kong SAR, China', 'Hong Kong')
gdp['Country Name'] = gdp['Country Name'].replace('Korea, Rep.', 'Korea, South')
gdp['Country Name'] = gdp['Country Name'].replace('Syrian Arab Republic', 'Syria')
gdp['Country Name'] = gdp['Country Name'].replace('Trinidad and Tobago', 'Trinidad & Tobago')
gdp['Country Name'] = gdp['Country Name'].replace('Russian Federation', 'Russia')
gdp['Country Name'] = gdp['Country Name'].replace('Iran, Islamic Rep.', 'Iran')
gdp['Country Name'] = gdp['Country Name'].replace('Egypt, Arab Rep.', 'Egypt')
gdp['Country Name'] = gdp['Country Name'].replace('North Macedonia', 'Macedonia (North Macedonia)')
gdp['Country Name'] = gdp['Country Name'].replace('Cabo Verde', 'Cape Verde')
gdp['Country Name'] = gdp['Country Name'].replace('Papua New Guinea', 'Pap. New Guinea')

efi_countries = pd.unique(efi_raw['Countries'])
gdp_countries = pd.unique(gdp['Country Name'])
idx1 = pd.Index(efi_countries)
idx2 = pd.Index(gdp_countries)
print(idx1.difference(idx2))
```

```python
countryTable = pd.DataFrame({ 'Country': efi_countries })
countryTable.to_csv('region_template.csv')
regions = pd.read_csv('regions.csv')
regions.head()
```

```python
gdps_by_year = pd.DataFrame({
    'Country': [],
    'Year': [],
    'GDP': []
})
for row in gdp.iterrows():
    data = row[1]
    country = data['Country Name']
    years = data.drop(['Country Name'])
    for year, value in zip(years.keys(), years):
        gdps_by_year = gdps_by_year.append({
            'Country': country,
            'Year': year,
            'GDP': value
        }, ignore_index=True)
        
```

```python
gdps_by_year['Year'] = gdps_by_year['Year'].astype('int64')
```

```python
gdps_by_year.head()
```

```python
efi = efi_raw[['Year', 'Countries', 'Economic Freedom Summary Index', 'Size of Government', 'Legal System & Property Rights', 'Sound Money', 'Freedom to Trade Internationally', 'Regulation']]
efi = efi.rename(columns={'Countries': 'Country'})
efi.head(1)
```

```python
table = pd.merge(gdps_by_year, efi)
table = pd.merge(table, regions)
table.iloc[35:38]
```

```python
table.to_csv('../public/data/data.csv')
```
